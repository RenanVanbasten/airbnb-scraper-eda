```python
import pandas as pd

df_bruto = pd.read_csv('../data/airbnb_listings.csv')
df_bruto
```

```python
df_limpo = df_bruto.copy()
df_limpo['preco_noite_numerico'] = df_limpo['preco_noite'].str.replace('R$', '', regex=False).astype(float)
df_limpo['nota_avaliacao'] = df_limpo['avaliacao'].str.split(' ').str[0].str.replace(',', '.').astype(float)
df_limpo['numero_avaliacoes'] = df_limpo['avaliacao'].str.split(' ').str[1].astype(int)
df_limpo
```
